```python
import json
import pandas as pd

with open('data_file.json', 'r') as f:
    data = json.load(f)

# Extracting information from the JSON data (same as before)
total_hits = data.get("totalHits", "N/A")
current_page = data.get("currentPage", "N/A")
total_pages = data.get("totalPages", "N/A")
query = data["foodSearchCriteria"].get("query", "N/A")
foods = data.get("foods", [])

# Define a list of nutritional components to extract (same as before)
nutritional_components = [
    "Protein",
    "Total lipid (fat)",
    "Carbohydrate, by difference",
    "Energy",
    "Sugars, total including NLEA",
    "Fiber, total dietary",
    "Calcium, Ca",
    "Iron, Fe",
    "Sodium, Na",
    "Vitamin A, IU",
    "Vitamin C, total ascorbic acid",
    "Cholesterol",
    "Fatty acids, total trans",
    "Fatty acids, total saturated",
    "Fatty acids, total monounsaturated",
    "Fatty acids, total polyunsaturated",
]

# Create a list to store the extracted data
data_list = []

# Loop through the list of foods and extract nutritional information for each item
for food in foods:
    fdc_id = food.get("fdcId", "N/A")
    description = food.get("description", "N/A")
    dataType = food.get("dataType", "N/A")
    brandName = food.get("brandName", "N/A")
    ingredients = food.get("ingredients", "N/A")

    # Extract serving size information
    serving_size = food.get("servingSize", "N/A")
    serving_size_unit = food.get("servingSizeUnit", "N/A")

    # Extract nutritional information using .get
    food_nutrients = food.get("foodNutrients", {})
    
    # Create a dictionary to store nutritional information
    nutritional_info = {component: "N/A" for component in nutritional_components}
    
    for nutrient in food_nutrients:
        nutrient_name = nutrient.get("nutrientName", "N/A")
        nutrient_value = nutrient.get("value", "N/A")
        nutritional_info[nutrient_name] = nutrient_value

    # Calculate energy per gram
    energy = nutritional_info.get("Energy", "N/A")
    if serving_size_unit == "g" and energy != "N/A":
        energy_per_gram = float(energy) / serving_size
        nutritional_info["Energy per Gram"] = energy_per_gram
    else:
        nutritional_info["Energy per Gram"] = "N/A"

    # Append the extracted data to the list
    data_list.append({
        "FDC ID": fdc_id,
        "Description": description,
        "Data Type": dataType,
        "Brand Name": brandName,
        "Ingredients": ingredients,
        "Serving Size": serving_size,
        "Serving Size Unit": serving_size_unit,
        **nutritional_info  # Include nutritional data in the dictionary
    })

# Create a DataFrame from the list of data
df = pd.DataFrame(data_list)

# Display the DataFrame
# df.head()  # You can also use df.head(n) to display the first n rows
```

```python
# Clean up DataFrame

# Rename "Carbohydrate, by difference" to Carbohydrate
df.rename(columns={'Carbohydrate, by difference': 'Carbohydrate'}, inplace=True)
# Rename "Total lipid (fat)" to Fat
df.rename(columns={'Total lipid (fat)': 'Fat'}, inplace=True)
# Rename "Sugars, total including NLEA" to Total_Sugars
df.rename(columns={'Sugars, total including NLEA': 'Total_Sugars'}, inplace=True)

# Only keep columns we want
columns_to_keep = [
    'Description',
    'Carbohydrate',
    'Fat',
    'Total_Sugars',
    'Protein',
    'Energy',
    'Fiber, total dietary',
    'Cholesterol',
    'Fatty acids, total saturated',
    'Caffeine',
    'Lactose',
    'Serving Size',
    'Energy per Gram'
]

df = df[columns_to_keep]
```
